```python
filepath = '/Users/leoalonso/Desktop/Python/Hospital1.txt'
df = pd.read_csv(filepath)
```

```python
print(df)
```

```python
# Calculating statistics
num_readmitted = np.sum(df[' Readmission'])
satisfaction_staff = np.mean (df[' StaffSatisfaction'])
satisfaction_cleanliness = np.mean (df[' CleanlinessSatisfaction'])
satisfaction_food = np.mean (df[' FoodSatisfaction'])
satisfaction_comfort = np.mean (df[' ComfortSatisfaction'])
satisfaction_communication = np.mean (df[' CommunicationSatisfaction'])
```

```python
# Printing out descriptive statistics
print(f"Number of patients readmitted: {num_readmitted}.")
print(f"Average staff satisfaction: {satisfaction_staff}.")
print(f"Average cleanliness satisfaction: {satisfaction_cleanliness}.")
print(f"Average food satisfaction: {satisfaction_food}.")
print(f"Average comfort satisfaction: {satisfaction_comfort}.")
print(f"Average communication satisfaction: {satisfaction_communication}.")
```

```python
# Calculate overall satisfaction
df['OverallSatisfaction'] = df[[' StaffSatisfaction', ' CleanlinessSatisfaction', ' FoodSatisfaction', ' ComfortSatisfaction', ' CommunicationSatisfaction']].mean(axis=1)
plt.boxplot(df['OverallSatisfaction'], showfliers=True)
```

```python
# Logistic regression
X = df['OverallSatisfaction'].values.reshape(-1, 1)
Y = df[' Readmission']

log_reg = LogisticRegression().fit(X, Y)
```

```python
# Correlation results
correlation_coefficient = log_reg.coef_[0][0]

if correlation_coefficient > 0:
    print("Logistic regression results indicated a: ")
    if correlation_coefficient > 0.5:
        print("Moderate correlation")
    elif correlation_coefficient > 0.7:
        print("Strong correlation")
    else:
        print("Weak correlation")
else: 
    print("Logistic regression results indicated: ")
    print("No correlation")

print(f"The correlation coefficient was {correlation_coefficient}")
```

```python
# plot the data
plt.scatter(X, Y)
plt.xlabel('Overall Satisfaction Scores')
plt.ylabel('Readmission')
plt.title('Logistic Regression - Overall Satisfaction vs Readmission')
plt.plot(X, log_reg.predict(X), color='blue')
plt.xlim(2, 5)

plt.show()
```

```python
def compare_correlation(hospital1_coefficient, hospital2_coefficient):
    """
    Compare correlation coefficients of two hospitals.

    Parameters:
    - hospital1_coefficient: Correlation coefficient for hospital 1.
    - hospital2_coefficient: Correlation coefficient for hospital 2.

    Returns:
    - comparison_result: A string describing the comparison result.
    """

    # Define a threshold for considering the correlation strength
    threshold = 0.5  # You can adjust this threshold based on your criteria

    # Compare correlation coefficients
    if abs(hospital1_coefficient) > threshold and abs(hospital2_coefficient) > threshold:
        comparison_result = "Both hospitals have strong correlation."
    elif abs(hospital1_coefficient) > abs(hospital2_coefficient):
        comparison_result = "Hospital 1 may be doing slightly better in terms of patient satisfaction and readmission rates compared to Hospital 2."
    elif abs(hospital2_coefficient) > abs(hospital1_coefficient):
        comparison_result = "Hospital 2 may be doing slightly better in terms of patient satisfaction and readmission rates compared to Hospital 1."
    else:
        comparison_result = "Both hospitals have similar weak correlation."

    return comparison_result

# Correlation coefficients for hospitals
hospital1_coefficient = 0.3912720623165687
hospital2_coefficient = 0.33034401828974597

# Compare correlation coefficients
result = compare_correlation(hospital1_coefficient, hospital2_coefficient)

print(result)
```
